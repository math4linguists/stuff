Пока мы решили искать нужные нам фрагменты при помощи регулярных выражений.

```python
import re
```

```python
# Ищем div, который обрамляет информацию о персоне.
profs = re.findall(r'<div class="fa-person__box">.+?</div>', res.text, re.S)
```

```python
# Выделяем из него гиперссылки на персональные страницы сотрудников школы лингвистики.
hrefs = [re.findall(r'href="(.+?)"', r, re.S)[0][2:] for r in profs]
```

```python
# Обходим полученные ссылки, переходим на страницы сотрудников,
# получаем со страницы ФИО сотрудника (а захотим - так и всё остальное).
for url in hrefs:
    res2 = requests.get("https://"+url)
    print(re.findall('<h1.*?>(.+?)</h1>', res2.text))
```

```python
# Страницы откликаются только когда к ним добавляют названия протокола доступа.
requests.get("https://"+hrefs[0])
```

При помощи библиотеки natasha можно извлекать именованные сущности. Создаем NamesExtractor (для извлечения имен) и с его помощью извлечи имена в документ JSON.

```python
import natasha
```

```python
from natasha import NamesExtractor
from natasha.markup import show_markup, show_json
```

```python
extractor = NamesExtractor()

text = '''
Благодарственное письмо   Хочу поблагодарить учителей моего, теперь уже бывшего, одиннадцатиклассника:  Бушуева Вячеслава Владимировича и Бушуеву Веру Константиновну. Они вовлекали сына в интересные внеурочные занятия, связанные с театром и походами.

Благодарю прекрасного учителя 1"А" класса - Волкову Наталью Николаевну, нашего наставника, тьютора - Ларису Ивановну, за огромнейший труд, чуткое отношение к детям, взаимопонимание! Огромное спасибо!
'''
# Извлекаем имена из текста
matches = extractor(text)
spans = [_.span for _ in matches]
# Формируем результат в json.
facts = [_.fact.as_json for _ in matches]
show_markup(text, spans)
# Вывод json.
show_json(facts)
```
